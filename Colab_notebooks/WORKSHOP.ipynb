```
# Coding Challenge-1 : Find the Largest Number

number_count = int(input("How many numbers will you enter: "))
numbers = []
for i in range(number_count):
  number = int(input("Please enter the number: "))
  numbers.append(number)
max_number = max(numbers)  
print("The largest number is :", max_number)
```

```
# Coding Challenge-2 : Group Anagrams

given_l = ["eat", "tea", "tan", "ate", "nat", "bat","tab","tana","anat"]
char = {}
value_l =[]
word_l = []
for i in given_l:
  char[i] = sorted(i)
  if sorted(i) not in value_l:
    value_l.append(sorted(i))
for j in value_l:
  word_str = ""
  for i in given_l:
    if j==char[i]:
      word_str += i +" "
  word_str = word_str.strip().split(" ")
  word_l.append(word_str)      
print(word_l)
```

```
test = " "
lis_1 = []
while test != "ok":
  test = input("Type 'ok' when you are done: ").lower()
  if test != "ok":
    lis_1.append(int(test))
el = 0
for i in range(1,len(lis_1)-1):
   right_max = lis_1[i]
   left_max = lis_1[i]
   for j in range(i):
     if lis_1[j] > left_max:     #find the max element of i in the left
        left_max = lis_1[j]
   for j in range(i+1,len(lis_1)):
      if lis_1[j] > right_max:
        right_max = lis_1[j] 
   el += min(right_max,left_max)-lis_1[i]    
print("Rain-trapped area :", el,)   
```
