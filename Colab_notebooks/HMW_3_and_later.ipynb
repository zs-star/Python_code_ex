```python
numbers = [1, 3, 7, 4, 3, 0, 3, 6, 3]
repeat_max = 0;
for i in numbers:
  repeat = 0
  for j in numbers:
    if i==j:
        repeat += 1
    else:
        repeat += 0 
  if repeat > repeat_max:     
    repeat_max = repeat 
    max_value = i    
print("The most frequent number is {} and it was {} times repeated.".format(max_value,repeat_max))
```

```python
# Find out if a given number is an "Armstrong Number".
number = input("Enter the positive number: ")
if number.isdigit():
  if int(number) >= 0:
      number = int(number)
      print("Let's check if it is Armstrong number or not!")
      length_no = len(str(number))
      sum_no = 0   
      for i in str(number):   
         sum_no += int(i)**length_no 
      if sum_no==number:
         print("{} is an Armstrong number".format(number)) 
      else:
         print("{} is not Armstrong number".format(number)) 
  else:
      print(" It is an invalid entry. Don't use non-numeric, float, or negative values!") 
      
else:
      print(" It is an invalid entry. Don't use non-numeric, float, or negative values!")  
```

```python
while True :
    number = input("enter a positive number.")
    digits = len(number)
    summ = 0
    if not number.isdigit() :
        print(number, "is invalid entry. enter numberic value!")
    elif int(number) >= 0 :
        for i in range(digits) :
            summ = summ + int(number[i]) ** digits
        if summ == int(number) :
            print(number, "is an Armstrong Number.")
            break
        else :
            print(number, "is not an Armstrong Number.")
            break
```

```python
# Task : Write a program that takes a number from the user and prints the result to check if it is a prime number.
number = input("Enter the number: ")
if number.isdigit():
  if int(number)>=1:
    number = int(number)      
    count = 0
    for i in range(1,number):
      if number%i==0:
        count+=1
    if count>1:
      print("Output: {} is not a prime number".format(number))
    else:
      print("Output: {} is a prime number".format(number)) 
else:
    print("Enter the valid  number") 
```

```python
n = int(input("Enter a number to check if it is a prime number."))
count = 0
for i in range(1, n+1) :
    if n % i == 0 :
        count += 1
if (n == 0) or (n == 1) or (count >=3) :
    print(n, "is not a prime number.")
else:
    print(n, "is a prime number")
```
