```
score = int(input("Enter your score:"))
if score >= 90:
  if score>=95:
    print("Your score is : A+")
  else: 
    print("Your score is : A")  
elif score >= 80:
  if score>=85:
    print("Your score is : B+")
  else: 
    print("Your score is : B") 
else :
    print("Your score is : B- or below")    
```
