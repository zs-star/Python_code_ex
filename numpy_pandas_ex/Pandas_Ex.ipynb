```python
def my_list(lis):
    mul,new=1,[]
    for i in lis:
        mul*=i
    for i in lis:
        try:
            new.append(mul//i)
        except:
              print("Zero Division Error") 
              break  
    if new:
        return new
my_list([3, 2, 1])
```

```python
def my_list(lis):
    new=[]
    for i in lis:
        mul=1
        for j in lis:
            if i!=j:
                mul*=j
        new.append(mul)
    return new  
my_list([1, 2, 3, 4, 5])
```
